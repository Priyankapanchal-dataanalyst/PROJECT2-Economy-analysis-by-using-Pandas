```python
import matplotlib.pyplot as plt #visualizationimport 
import  seaborn as sns  #distribution visualisation
# seaborn are used for advance visualization e.x -->distribution plot, line plot
%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2

import warnings 
warnings.filterwarnings('ignore') #os error
```

```python
df.head()
```

```python
df["InternetUsers"]
```

```python
vis1 = sns.distplot(df["InternetUsers"])  #Univariate Analysis statistics
plt.show()
```

```python
vis2 = sns.displot(df["InternetUsers"])
plt.show()
```

```python
vis3 = sns.distplot(df["InternetUsers"],bins = 10)
plt.show()
```

```python
vis4 = sns.boxplot(data = df, x = "IncomeGroup", y = "BirthRate") #BOXPLOTS #BIVARIATE bivariable ANALYSIS
plt.show()
```

```python
vis5= sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate') 
plt.show()
```

```python
vis6= sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False) 
plt.show()
```

```python
vis7= sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True)  # if you want line then use fit_reg = true
plt.show()
```

```python
vis8= sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False,hue = 'IncomeGroup')  # if you want line then use fit_reg = true
plt.show()
```

```python
vis9= sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True,hue = 'IncomeGroup')  # if you want line then use fit_reg = true
plt.show()
```
